```python
students = {
    "Alice" : {"maths":78, "science":74, "social":80},
    "Bob" : {"maths":67, "science":70, "social":60},
    "Charlie": {"maths":60, "science":70, "social":80},
    "David" : {"maths":54, "science":61, "social":69},
    "Eiffel" : {"maths":81, "science":85, "social":90}
}

for name, subjects in students.items():
    total = sum(subjects.values())
    avg = total / len(subjects)
    subjects['total'] = total
    subjects['average'] = avg


student_list = list(students.items())


def get_total_marks(item):
    return item[1]['total']


student_list.sort(key=get_total_marks, reverse=True)


ranks = {}
current_rank = 1
previous_total = None
previous_rank_assigned = 0 

for name, data in student_list:
    if previous_total is None or data['total'] < previous_total:
        ranks[name] = current_rank
        previous_rank_assigned = current_rank 
    else:
        ranks[name] = previous_rank_assigned

    previous_total = data['total']
    current_rank += 1 

print("---  Ranks ---")
for name, details in students.items():
    details['rank'] = ranks[name]
    print(f"Name: {name}, Total Marks: {details['total']}, Average Marks: {details['average']:.2f}, Rank: {details['rank']}")
```

```python
courses = [
    ("C101", "Python Programming", 9000),
    ("C102", "Java Programming", 4000),
    ("C103", "DevOps Programming", 12000),
    ("C104", "Dotnet Programming", 5000),
    ("C105", "Machine Learning Programming", 19000)
]

def display_courses():
    print("\nThe list of courses available:")
    for code, title, fee in courses:
        print(f"{code} -- {title} -- {fee}")

def get_valid_courses():
    registered_courses = set()
    course_list = []
    while True:
        display_courses()
        choice = input("Enter the course code - 'DONE' to finish: ").strip()
        if choice.upper() == 'DONE':
            break
        if choice in registered_courses:
            print("Course already registered!")
            continue
        found = False
        for course in courses:
            if course[0] == choice:
                course_list.append(course)
                registered_courses.add(choice)
                print("Course Added Successfully...!")
                found = True
                break
        if not found:
            print("Invalid Course Code")
    return course_list

def register_students():
    students = []
    while True:
        try:
            n = int(input("Enter number of students: "))
            if n < 1:
                print("Number of students must be at least 1.")
                continue
            break
        except ValueError:
            print("Invalid input! Please enter a valid integer.")
    
    for _ in range(n):
        name = input("\nEnter the student name: ").strip()
        sid = input("Enter the student id: ").strip()
        course_list = get_valid_courses()
        total_fees = sum(course[2] for course in course_list)
        student = {
            "id": sid,
            "name": name,
            "courses": course_list,
            "total_fees": total_fees
        }
        students.append(student)
    return students

def display_students(students):
    for s in students:
        print("\nStudent Id:", s['id'])
        print("Name of Student:", s['name'])
        print("Registered Courses:")
        for course in s['courses']:
            print(f"{course[0]} -- {course[1]} -- {course[2]}")
        print("Total Fee to pay:", s['total_fees'])

# Main execution
students = register_students()
display_students(students)
```
